Desafio:

Faça um programa que receba 4 notas de um aluno e faça:

- Média Ponderada sabendo que cada nota tem um peso que deve ter informado pelo usuário
- Exiba o resultado na tela

media = (nota1 * peso1) + (nota2 * peso2) ... / (peso1 + peso2...)


```python
notaA = float(input("Informe a Nota A: "))
pesoA = int(input("Informe o peso da Nota A: "))
notaB = float(input("Informe a Nota B: "))
pesoB = int(input("Informe o peso da Nota B: "))
notaC = float(input("Informe a Nota C: "))
pesoC = int(input("Informe o peso da Nota C: "))
notaD = float(input("Informe a Nota D: "))
pesoD = int(input("Informe o peso da Nota D: "))

mediaPond = ((notaA*pesoA) + (notaB*pesoB) + (notaC*pesoC) + (notaD*pesoD)) / (pesoA+pesoB+pesoC+pesoD)

print(f"A média ponderada é: {round(mediaPond,2)}")
```

Atividade de Fixação

- Objetivo: Criar um sistema simples de cadastro de produtos.
- Descrição: Crie uma lista vazia chamada estoque . Em seguida, crie um dicionário para representar um produto com chaves nome , preco e
quantidade . Adicione este dicionário à lista estoque . Repita para mais dois produtos. Por fim, imprima o nome e o preço de todos os produtos no estoque.
- Contexto IA: Simula a organização de dados de um catálogo de produtos, que
pode ser usado para sistemas de recomendação.

```python
estoque = []

produto1 = {"nome":"Notebook", "preco":3500.00, "quantidade":10}
produto2 = {"nome":"Mouse", "preco":68.00, "quantidade":65}
produto3 = {"nome":"Teclado", "preco":187.50, "quantidade":48}

estoque.append(produto1)
estoque.append(produto2)
estoque.append(produto3)


```

```python
print("\n ----- Produtos em Estoque ------")

for produto in estoque:
  print(f"Nome: {produto['nome']}, Preço: R$ {produto['preco']:.2f}")
```
