#Bubble Sort = "Bolhas subindo na água"


Lista de números = [5, 1, 4, 2, 8]

1. Comparar 5 e 1 = [1,5,4,2,8]
2. Comparar 5 e 4 = [1,4,5,2,8]
3. Comparar 5 e 2 = [1, 4, 2, 5, 8]


```python
def bubble_sort(lista):
  n = len(lista)

  for i in range(n):
    for j in range(0, n - i - 1):
      if lista[j] > lista[j + 1]:
        lista[j], lista[j + 1] = lista[j + 1], lista[j]

  return lista
```

```python
minha_lista_bubble = [64, 34, 25, 12, 22, 11, 90]
print(f"Lista Original = {minha_lista_bubble} \n")

lista_ordenada_bubble = bubble_sort(minha_lista_bubble)
print(f"Lista Ordenada com Bubble Sort = {lista_ordenada_bubble}")
```

#Selection Sort = Seleciona o Menor e Coloca no lugar Certo!

Lista = [5, 1, 4, 2, 8]
1. [1, 5,4,2,8]
2. [1,2,4,5,8]

```python
def selection_sort(lista):
  n = len(lista)

  for i in range(n):
    idx_min = i

    for j in range(i + 1, n):
      if lista[j] < lista[idx_min]:
        idx_min = j

    lista[i], lista[idx_min] = lista[idx_min], lista[i]

  return lista
```

```python
minha_lista_selection = [64,25,12,22,11]
print(f"Lista Original = {minha_lista_selection} \n\n")
lista_ordenada_selection = selection_sort(minha_lista_selection)
print(f"Lista Ordenada Selecion Sort = {lista_ordenada_selection}")
```

#Insertion Sort = Organizando Cartas de um Baralho

Lista = [5,1,4,2,8]

1. [1,5]
2. [1,4,5]
3. [1,2,4,5]
4. [1,2,4,5,8]


```python
def insertion_sort(lista):
  for i in range(1, len(lista)):
    chave = lista[i]

    j = i - 1
    while j >= 0 and chave < lista[j]:
      lista[j + 1] = lista[j]

      j -= 1

    lista[j + 1] = chave

  return lista
```

```python
minha_lista_insertion = [12,11,13,5,6,1]
print(f"Lista Original = {minha_lista_insertion} \n\n")
lista_ordenada_insertion = insertion_sort(minha_lista_insertion)
print(f"Lista Ordenada Insertion Sort = {lista_ordenada_insertion}")
```

 1. [38,27,43,3] [9,82,10]

```python
def merge_sort(lista):
  if len(lista) <= 1:
    return lista

  meio = len(lista) // 2
  esquerda = lista[:meio]
  direita = lista[meio:]

  esquerda_ordenada = merge_sort(esquerda)
  direita_ordenada = merge_sort(direita)

  return merge(esquerda_ordenada, direita_ordenada)
```

```python
def merge(esquerda, direita):
   resultado = []

   i = j = 0

   while i < len(esquerda) and j < len(direita):
    if esquerda[i] < direita[j]:
      resultado.append(esquerda[i])
      i += 1
    else:
      resultado.append(direita[j])
      j += 1

   resultado.extend(esquerda[i:])
   resultado.extend(direita[j:])

   return resultado
```

```python
minha_lista_merge = [38,27,43,3,9,82,10]
print(f"Lista Original = {minha_lista_merge} \n")
lista_ordenada_merge = merge_sort(minha_lista_merge)
print(f"Lista Ordenad [Merge Sort] = {lista_ordenada_merge}")
```

```python
class Fila:
  def __init__(self):
    self.max_tamanho = 25

    self.elementos = [None] * self.max_tamanho

    self.inicio = 0
    self.fim = 0
    self.quantidade = 0

  def fila_vazia(self):
    return self.quantidade == 0

  def cheia(self):
    return self.quantidade == self.max_tamanho

  def adicionar(self, nome):
    if self.cheia():
      print("A fila está Cheia. Não é possível adicionar mais elementos...")
      return

    self.elementos[self.fim] = nome

    self.fim = (self.fim + 1) % self.max_tamanho
    self.quantidade += 1
    print(f"{nome} foi adicionado à fila!")

  def remover(self):
    if self.fila_vazia():
      print("A fila está Vazia. Não há elementos para remover.")
      return None

    nome = self.elementos[self.inicio]
    self.elementos[self.inicio] = None

    self.inicio = (self.inicio + 1) % self.max_tamanho
    self.quantidade -= 1
    print(f"{nome} foi removido da Fila.")
    return nome


  def listar(self):
    if self.fila_vazia():
      print("A fila está Vazia.")
      return

    print("\n Elementos da Fila: ")
    i = self.inicio
    for _ in range(self.quantidade):
      print(f"- {self.elementos[i]}")
      i = (i + 1) % self.max_tamanho


  def limpar(self):
    while not self.fila_vazia():
      self.remover()
    print("A fila foi esvaziada!")




```
