# **Atividade 1**

Dada a lista L = [5, 7, 2, 9, 4, 1, 3], escreva um programa que imprima as seguintes informações: a) tamanho da lista; b) maior valor da lista; c) menor valor da lista; d) soma de todos os elementos da lista; e) lista em ordem crescente; f) lista em ordem decrescente.

```python
#lista
L = [5, 7, 2, 9, 4, 1, 3]

# a) tamanho da lista
tamanho = len(L)

# b) maior valor da lista
maior = max(L)

# c) menor valor da lista
menor = min(L)

# d) soma de todas os elementos da lista
soma = sum(L)

# e) lista em ordem crescente
crescente = sorted(L)

# f) lista em ordem decrescente
decrescente = sorted(L, reverse=True)

print("a) tamanho", tamanho) # O len tem a função de conta quantos elementos tem na lista.
print("b) maior", maior) # O max conta o maior elemento da lista
print("c) menor", menor) # O min conta o menor elemento da lista
print("d) soma", soma)  # O sum soma todos os elementos da lsita
print("e) crescente", crescente) # O sorted organiza os números do menor para o maior
print("f) decrescente", decrescente) # O sorted usando o reverse=True faz a ordenação do maior para o menor
```

# **Atividade 2**

Faça um programa que leia 4 valores, calcule a média e imprima positivo ou negativo (para ser posi tivo a média deve ser acima de 1).

```python

# ler 4 valores
v1 = float(input("Digite o primeiro valor: "))  # As variaveis v1, v2, v3, v4, recebe ( = ) ou atribuir ( = ), os valores digitados.
v2 = float(input("Digite o segundo valor: "))   # O float é usado pra converter o número digitado que é uma string em números decimais (números quebrados). O input é uma função que serve pra receber dados digitados do usuário, so recebe string
v3 = float(input("Digite o terceiro valor: "))  # e "Digite o valor" vai ser os números que você vai mencionar
v4 = float(input("Digite o quarto valor: "))

# calcular a média
media = (v1 + v2 + v3 + v4) /4 # A variavel que é chamada (média) recebe ( = ) ou atrubuir ( = ). E calcula a média dos quatro valores (v1 + v2 + v3 + v4) /4 e depois dividir por 4.

# verificar se é positivo ou negativo
if media > 1:
  print("Positivo") # if media > 1 Verifica se a média é maior que 1.
else:
  print("Negativo")  # else se não for maior que 1, séra considerado negativo.

#Mostra a Média
print("Média", media)
```

# **Atividade 3**

Escreva um programa que leia 20 valores inteiros e informe a média deles, o maior e o menor valor.

```python
# Atividade 3

numeros = [1, 3, 5, 6, 7, 8, 10, 13, 14, 15, 17, 18, 20, 21, 23, 25, 26, 28, 30, 34] #numeros = [...] → aqui colocamos os 20 valores inteiros.

#Média
media = sum(numeros) / 20 #sum(numeros) → soma todos os números.
                          #20 → conta quantos números existem na lista (20).
#Maior e menor valor
maior = max(numeros) #max(numeros) → encontra o maior valor.
menor = min(numeros) #min(numeros) → encontra o menor valor.

#mostrar resultados
print("Média: ", media) # a média é aonde que fica entre o meio dos números, no caso os números são (15 e 17).
print("maior: ", maior) # é o maior número que tem nesta lista.
print("menor: ", menor) # é o menor número que tem na lista.
```

# **Atividade 4**

Crie uma função para desenhar uma linha, usando o caractere '_' (underline). O tamanho da linha deve ser definido na chamada da função.

```python

def desenhar_linha(tamanho): #def desenhar_linha(tamanho): → cria uma função chamada desenhar_linha. Tamanho é o valor que você escolhe quando chama a função.
   print("_" * tamanho) #"_" * tamanho → repete o caractere _ várias vezes.
                         #print() → mostra a linha na tela.
# Exemplo de chamada da função
desenhar_linha(10)
desenhar_linha(20)
```

```python
#aqui é outro exemplo que você pode adicionar, fazer um desenho de uma casa e etc
#usando caracteres diferentes como o "/", "|", "\\","_", " " transformando um formato de uma casa, usando a criatividade

def desenhar_casa(tamanho):
                #(tamanho) é a quantidade de itens dentro dela.
                # Quando você chama desenhar_casa(5), o valor de tamanho será 5.
    # Telhado
    for i in range(1, tamanho + 1):  # O range é usado para gerar uma sequência de números. usando o lacos
        print(" " * (tamanho - i) + "/" + " " * (i * 2 - 2) + "\\")  #Esse for repete o código várias vezes para formar o triângulo do telhado.

    # Base do telhado
    print("/" + "_" * (tamanho * 2 - 2) + "\\")

    # Corpo da casa
    for i in range(tamanho):
        print("|" + " " * (tamanho * 2 - 2) + "|")

    # Chão da casa
    print("|" + "_" * (tamanho * 2 - 2) + "|")

# Exemplo
desenhar_casa(5) # È o tamanho da casa.
```

# **Atividade 5**

Crie uma função que receba como parâmetro uma lista, com valores de qualquer tipo. A função deve imprimir todos os elementos da lista, enumerando-os.

```python

def imprimir_lista(lista): #A função imprimir_lista(lista) recebe uma lista como parâmetro.
    for i, elemento in enumerate(lista, start=1): #O enumerate() serve para pegar o número da posição e o elemento ao mesmo tempo. #start=1 faz a contagem começar em 1 (em vez de 0).
        print(f"{i} - {elemento}") #O print(f"{i} - {elemento}") mostra o número e o valor da lista.

# Exemplo de uso
minha_lista = ["Python", 10, 3.5, True, "Olá"] #O elemento da lista
imprimir_lista(minha_lista)
```

```python

# Programa para converter horas em segundos

horas = float(input("Digite a quantidade de horas: ")) ## O float é usado pra converter o número digitado que é uma string em números decimais. E o input() captura o valor digitado pelo usuário. float() permite que o usuário digite horas com número decimal (ex: 1.5 horas).
                        #Sabemos que 1 hora = 3600 segundos, então multiplicamos:
segundos = horas * 3600 #segundos = horas * 3600

print("Isso equivale a", segundos, "segundos.") #Depois o resultado é mostrado com print().
```
