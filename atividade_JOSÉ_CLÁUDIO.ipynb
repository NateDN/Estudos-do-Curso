# **Atividade 1**

```python
#lista
L = [5, 7, 2, 9, 4, 1, 3]

# a) tamanho da lista
tamanho = len(L)

# b) maior valor da lista
maior = max(L)

# c) menor valor da lista
menor = min(L)

# d) soma de todas os elementos da lista
soma = sum(L)

# e) lista em ordem crescente
crescente = sorted(L)

# f) lista em ordem decrescente
decrescente = sorted(L, reverse=True)

print("a) tamanho", tamanho) # O len tem a função de conta quantos elementos tem na lista.
print("b) maior", maior) # O max conta o maior elemento da lista
print("c) menor", menor) # O min conta o menor elemento da lista
print("d) soma", soma)  # O sum soma todos os elementos da lsita
print("e) crescente", crescente) # O sorted organiza os números do menor para o maior
print("f) decrescente", decrescente) # O sorted usando o reverse=True faz a ordenação do maior para o menor
```

# **Atividade 2**

```python

 # O input() serve para ler dados digitados pelo usuário.
 # O float() serve para transforma o valor digitado em números com decimal.
 # (v1 + v2 + v3 + v4) /4 calcula a média dos quatro valores.
 # if media > 1 Verifica se a média é maior que 1.
 # else se não for maior que 1, séra considerado negativo.


# ler 4 valores
v1 = float(input("Digite o primeiro valor: "))
v2 = float(input("Digite o segundo valor: "))
v3 = float(input("Digite o terceiro valor: "))
v4 = float(input("Digite o quarto valor: "))

# calcular a média
media = (v1 + v2 + v3 + v4) /4

# verificar se é positivo ou negativo
if media > 1:
  print("Positivo")
else:
  print("Negativo")

#Mostrar a Média
print("Média", media)
```

# **Atividade 3**

```python
# Atividade 3

#numeros = [...] → aqui colocamos os 20 valores inteiros.
#sum(numeros) → soma todos os números.
#20 → conta quantos números existem na lista (20).
#sum(numeros) / 20 → calcula a média.
#max(numeros) → encontra o maior valor.
#min(numeros) → encontra o menor valor.

numeros = [1, 3, 5, 6, 7, 8, 10, 13, 14, 15, 17, 18, 20, 21, 23, 25, 26, 28, 30, 34]

#Média
media = sum(numeros) / 20

#Maior e menor valor
maior = max(numeros)
menor = min(numeros)

#mostrar resultados
print("Média: ", media)
print("maior: ", maior)
print("menor: ", menor)
```

```python

#def desenhar_linha(tamanho): → cria uma função chamada desenhar_linha.
#tamanho é o valor que você escolhe quando chama a função.
#"_" * tamanho → repete o caractere _ várias vezes.
#print() → mostra a linha na tela.

def desenhar_linha(tamanho):
   print("_" * tamanho)

# Exemplo de chamada da função
desenhar_linha(10)
desenhar_linha(20)
```

```python
#aqui é outro exemplo que você pode adicionar, fazer um desenho de uma casa e etc
#usando caracteres diferentes como o "/", "|", "\\","_", " " transformando um formato de uma casa, usando a criatividade

def desenhar_casa(tamanho):
                #(tamanho) é o número que controla o tamanho do desenho.
                # Quando você chama desenhar_casa(5), o valor de tamanho será 5.
    # Telhado
    for i in range(1, tamanho + 1):
        print(" " * (tamanho - i) + "/" + " " * (i * 2 - 2) + "\\")  #Esse for repete o código várias vezes para formar o triângulo do telhado.

    # Base do telhado
    print("/" + "_" * (tamanho * 2 - 2) + "\\")

    # Corpo da casa
    for i in range(tamanho):
        print("|" + " " * (tamanho * 2 - 2) + "|")

    # Chão da casa
    print("|" + "_" * (tamanho * 2 - 2) + "|")

# Exemplo
desenhar_casa(5)
```

# **Atividade 5**

```python


#A função imprimir_lista(lista) recebe uma lista como parâmetro.
#O enumerate() serve para pegar o número da posição e o elemento ao mesmo tempo.
#start=1 faz a contagem começar em 1 (em vez de 0).
#O print(f"{i} - {elemento}") mostra o número e o valor da lista.

def imprimir_lista(lista):
    for i, elemento in enumerate(lista, start=1):
        print(f"{i} - {elemento}")

# Exemplo de uso
minha_lista = ["Python", 10, 3.5, True, "Olá"]
imprimir_lista(minha_lista)
```
